# World Happiness Report 2021 Data Analysis

## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
%matplotlib inline
```

## Importing dataset

```python
df = pd.read_csv('data/WHR/world-happiness-report-2021.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

#### As we don't need all the attribute columns defined, so we are taking some of the attribute as below:

```python
data_columns = ['Country name', 'Regional indicator', 'Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
```

```python
df = df[data_columns].copy()
df.head()
```

```python
happy_df = df.rename({'Country name': 'country_name', 'Regional indicator': 'regional_indicator', 'Ladder score': 'happiness_score', 'Logged GDP per capita': 'GDP_per_capita', 'Social support': 'social_support', 'Healthy life expectancy': 'health_life_expectancy', 'Freedom to make life choices': 'freedom_to_make_life_choices', 'Generosity': 'generosity', 'Perceptions of corruption': 'perceptions_of_corruption'}, axis=1)
```

```python
happy_df.head()
```

### Plot between happiness and GDP

```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'

sns.scatterplot(x = happy_df.happiness_score, y = happy_df.GDP_per_capita, hue = happy_df.regional_indicator, data = happy_df, s = 200)

plt.title('Plot between happiness score and GDP')
plt.xlabel('Happiness score')
plt.ylabel('GDP per capita')
plt.legend(loc = 'upper left', fontsize='10')
plt.show()
```

```python
gdp_region = happy_df.groupby('regional_indicator')['GDP_per_capita'].sum()
gdp_region
```

```python
plt.rcParams['figure.facecolor'] = '#FFE5B4'
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP per capita by region')
plt.ylabel('')
plt.show()
```

### Total countries in each region

```python
total_country = happy_df.groupby('regional_indicator')[['country_name']].count()
total_country = total_country.reset_index()
total_country
```

```python
happy_df['regional_indicator'].value_counts()
```

### Correlation map

```python
cor = happy_df.corr(method = 'pearson')
f, ax = plt.subplots(figsize = (10, 5))
plt.rcParams['figure.facecolor'] = '#FFE5B4'

sns.heatmap(cor, mask = np.zeros_like(cor, dtype=np.bool),
           cmap='Blues', square=True, ax = ax)
plt.show()
```

### Corruption in different regions

```python
corruption = happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption = corruption.reset_index()
corruption
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['figure.facecolor'] = '#FFE5B4'

plt.bar(corruption.regional_indicator, corruption.perceptions_of_corruption)
plt.title('Perception of Corruption in various Regions', fontsize = 15)
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption Index', fontsize=15)
plt.xticks(rotation=30, ha='right', fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

### Top & Bottom 10 happiest countries life expectancy

```python
top_10 = happy_df.head(10)
bottom_10 = happy_df.tail(10)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
plt.rcParams['figure.facecolor'] = '#FFE5B4'
plt.tight_layout(pad = 2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x = top_10.country_name, y = top_10.health_life_expectancy, ax = axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')

xlabels = bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x = bottom_10.country_name, y = bottom_10.health_life_expectancy, ax = axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')

plt.show()
```

### Freedom to make life choices vs happiness score


```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'

sns.scatterplot(x = happy_df.freedom_to_make_life_choices, y = happy_df.happiness_score, hue = happy_df.regional_indicator, s = 200)
plt.legend(loc = 'upper left', fontsize = '12')
plt.title('Freedom to make life choices vs happiness score')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score')
plt.show()
```

### Top 10 most corrupted countries

```python
country = happy_df.sort_values(by = 'perceptions_of_corruption').tail(10)
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.facecolor'] = '#FFE5B4'

plt.title('Countries with most Perception of Corruption')
plt.xlabel('Country', fontsize=13)
plt.ylabel('Corruption Index', fontsize=12)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.perceptions_of_corruption)
plt.show()
```

### Top 10 least corrupted countries

```python
country = happy_df.sort_values(by = 'perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
plt.title('Countries with least Perception of Corruption')
plt.xlabel('Country', fontsize=13)
plt.ylabel('Corruption Index', fontsize=12)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.perceptions_of_corruption)
plt.show()
```

### Corruption vs happiness

```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'

sns.scatterplot(x = happy_df.happiness_score, y = happy_df.perceptions_of_corruption, hue=happy_df.regional_indicator, s=200)
plt.legend(loc = 'lower left', fontsize= '14')
plt.xlabel('Happiness score')
plt.ylabel('Corruption')
plt.show()
```
